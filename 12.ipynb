```python
def run_scored(batsman_name):
    vk = df[df['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head().index[0]
```

```python
run_scored('V Kohli')
```
